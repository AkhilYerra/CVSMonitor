```python
import time
import datetime
import re
def cvs_Job():
    from discord_webhook import DiscordWebhook
    import selenium
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    discordwebhookurl = 'https://discord.com/api/webhooks/815301752726355983/1Bmj6SMW-pdSD_s69Ui8EvjcuFmpmkMpqszdft4mYFftj1YhRgOGaRtx1qp5UpGU53u0'
    chromedriver_location = "./chromedriver"
    option = webdriver.ChromeOptions()
    option.add_argument('headless')
    driver = webdriver.Chrome(chromedriver_location,options=option)
    driver.get('https://www.cvs.com/immunizations/covid-19-vaccine')
    driver.find_element_by_link_text("New Jersey").click()
    paraDiv = driver.find_element_by_xpath("//*[contains(text(), 'At this time, all appointments in New Jersey are booked. We’ll add more as they become available.')]")
    someDiv = paraDiv.find_element_by_xpath('..')
    textDiv = someDiv.find_element_by_xpath('..')
#     /immunizations/covid-19-vaccine.vaccine-status.CA.json?vaccineinfo
    innerText = textDiv.get_attribute('innerHTML')
                                       
    if ('hidden' not in innerText):
        driver.close() 
    else:
        
        initialwebhook = DiscordWebhook(url=discordwebhookurl, content='https://www.cvs.com/vaccine/intake/store/cvd-schedule?icid=coronavirus-lp-vaccine-al-statetool')
        response = initialwebhook.execute()
        cityTable = driver.find_element_by_xpath("//div[@data-url='/immunizations/covid-19-vaccine.vaccine-status.NJ.json?vaccineinfo']")
        table = cityTable.find_element_by_xpath(".//*")
        row = table.find_elements_by_xpath(".//tr")
        for eachRow in row:
            if 'Available' in eachRow.get_attribute('outerHTML'):
                availableRow = eachRow.get_attribute('outerHTML')
                cityWebHook = DiscordWebhook(url=discordwebhookurl, content=availableRow)
                response = cityWebHook.execute()
        print(datetime.datetime.now())
        driver.close()
```

```python
import schedule
schedule.every(15).seconds.do(cvs_Job)

while 1:
    schedule.run_pending()
```
